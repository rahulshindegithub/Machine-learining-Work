```python
def pc_calculation(x,no):
    pca={}
    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(x)
    l,x = np.linalg.eig((x_scaled.T @ x_scaled)/(x_scaled.shape[0]-1))
    for i in range(no):
        pc = x_scaled @ x[:,i]
        pca[i]=pc
    pca_df = pd.DataFrame(pca)
    return pca_df
```

```python
pc_calculation(x,5)
```
