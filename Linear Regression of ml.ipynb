```python
  ### let's create a function to created adjusted R squared
def adj_r2(x,y):
    r2 = lr.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

## LASSO

```python
lassocv = LassoCV(alphas=None,cv=5, max_iter=20000000, normalize=True)
lassocv.fit(x_train,y_train)
```

```python
lassocv.alpha_
```

```python
lasso = Lasso(alpha = lassocv.alpha_)
lasso.fit(x_train,y_train)
```

```python
lasso.score(x_test,y_test)
```

## Ridge 

```python
ridgecv = RidgeCV(alphas=(0.1, 1.0, 10.0),cv=10, normalize=True)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.score(x_test, y_test)
```

```python
ridgecv.alpha_
```

```python
np.random.uniform(0,10,50)
```

```python
ridge_lr = Ridge(alpha = ridgecv.alpha_)
ridge_lr.fit(x_train, y_train)
```

```python
ridge_lr.score(x_test, y_test)
```

## ElasticNet

```python
elastic = ElasticNetCV(alphas=None, cv=10)
elastic.fit(x_train, y_train)
```

```python
elastic.alpha_
```

```python
elastic.l1_ratio
```

```python
elastic_lr = ElasticNet(alpha=elastic.alpha_, l1_ratio = elastic.l1_ratio)
elastic_lr.fit(x_train,y_train)
```

```python
elastic_lr.score(x_test,y_test)
```
